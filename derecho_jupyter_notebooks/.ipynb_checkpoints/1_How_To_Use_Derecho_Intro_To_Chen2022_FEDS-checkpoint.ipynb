# So, You Want To Use The Derecho Supercomputer? 

  This guide will help set you up to explore the Fire Events Data Suite (**FEDS**) code and project files using the National Center for Atmospheric Research (**NCAR**) and the University Corporation for Atmospheric Research (**UCAR**) Supercomputer, **Derecho**, of NCAR-Wyoming Supercomputing Center (**NWSC**) located in Cheyenne, Wyoming sponsored and by the National Science Foundation (**NSF**).   


Before we can do anything cool, we should probably do the coolest thing first.   
**Connect to the Supercomputer!**

## These Are The First Steps to Connect to Derecho.

In a browser, enter your credentials here:    
https://jupyterhub.hpc.ucar.edu/   

And join Derecho from the dropdown.   
You may adjust some settings, importantly, extend your session time.  

and/or   
Join through SSH using **MobaXterm**   
Download for PC: [MobaXterm](https://mobaxterm.mobatek.net/download.html)  

At a **terminal window** enter your username@server host  

**NOTE:**  
copy/paste works differently in mobaXterm.  
Right click is both copy and paste.  
And you should get a popup to approve pasting multiple lines.  
When pasting multiple lines of code, you may have to press enter after the paste to complete the last command.  

```ssh -X username@derecho.hpc.ucar.edu```  

Enter your password and wait for DUO Authentication, should be quick.   
If DUO does not ping your device, force DUO by adding ",push" like: yourpassword,push   

````python
# ssh -X $USER@derecho.hpc.ucar.edu
````

Type 'yes' if asked.

Wait for DUO and Authenticate.  

You are now on the Derecho supercomputer.   
**Good Job!**

Use the SSH tunnel to connect and view in browser.   
This is great for working in a notebook.  

````python
module load conda  # an important step, load Conda  
conda env list   # shows available kernels  
conda activate derecho_py  # adjust this to the required kernel  
````

We notice our kernel has changed:  
```('nameofkernel')username```  the ()before our user name.   

Next, we Launch Jupyter on the compute node (no browser needed, but we will use one)  
This command may take a moment to run.  
There will be a lot of print output.   
We will copy the web address at the end of the output and paste it into our browser.  

````python
jupyter lab --no-browser --ip=0.0.0.0 --port=8888
````

We paste the provided web address into our browser and we should see our file structure in the Jupyterhub.  
What has been output are the keys/token to enter the server.  
The token will change so the web address provided will change every time we run this command and want to connect.  
  
**Note:**  
--port=8888; can be change to 8888-8899 if previous number doesn't work   
derecho3; the number depends on the login node number where you launch jupyter   
If the http link doesn't work, clear caches and try again 
  
OR  

Try this next option:       

There is a good chance the provided output links when entering in browser will return an error and do not work/connect.  
If this is the case do not worry!  
We need to create a tunnel.   
**Open a new tab/terminal** and log in to derecho and authenticate as before then enter the following:  

```ssh -L 8888:derecho3:8888 $username$@derecho.hpc.ucar.edu```

````python
# ssh -L 8888:derecho3:8888 $USER@derecho.hpc.ucar.edu
# ssh -L 8889:derecho3:8888 $USER@derecho.hpc.ucar.edu  # change the port if one is busy
````

Now enter this into your browser:  

```http://localhost:8888```  

You should have access now.   
If not, repeat the steps. 

## Kernels and Setting The Environment

The following commands need to be run from a terminal.  
**You should not try to run the kernel setup from a notebook.**  

We will now create a kernel using Conda and Mamba for our project environment.   
We use Conda or Mamba (Mamba is C++ version of Conda) for the Python environment to host the Juptyer Notebook Kernel.   
This gives us access to the modules, libraries (like Pandas or NumPy) and dependencies in our Juptyer Notebook .ipynb files.   
We load the Conda (Python environment) module and then create kernel named '**myderecho_py**'   
We load the notebook and we load the needed Python modules that will be available for import.   


### Setup The Kernels 

Set the <a id='kernelfolder'></a> **default folder** for your kernels.   
I've named mine *"condas"*   

```conda config --prepend envs_dirs /glade/work/$USER/condas```   

Now, activate the environment: 


````python
module load conda
conda create -n myderecho_py python=3.11 -y
conda activate myderecho_py
````

### Install The Modules    

What are the modules that we are using for myderecho_py kernel?  

```conda install -y```  installs the Conda environment   
```jupyterlab```    the browser-based way to view our notebook  
```ipykernel```   connects Conda to our notebook   
```numpy```      math, handles arrays   
```scipy```    more math, integration, optimization, stats    
```pandas```    Excel for Python and data science, CSV-style rows and columns     
```matplotlib```   plotting, drawing charts, lines, and maps for display     
```xarray```      labeled interface, handles multi-dimensional arrays, N-Dimensions     
```dask```        splits files, allows "Lazy Loading," small chunks of very large files       
```netcdf4```      Network Common Data Form, driver that lets Python read units, dates, coordinates inside the file   
```h5py```      the foundation of netcd4, treats large files like a Python Dictionary, slices, saving memory    
```cartopy```    map drawing, handles projections     
```shapely```     geometric shapes, calculates coordinates area of this specific region   
```pyproj```      calculates coordinates, translator for Coordinate Reference Systems   
```cftime```      handles non-standard calendars, leap years 360 day year     

````python
conda install -y \
jupyterlab \
ipykernel \
numpy scipy pandas matplotlib \
xarray dask netcdf4 h5py \  
cartopy shapely pyproj \
cftime
````

Make the kernel available as a choice in the menu of the notebook (the top left).

````python
cd  
python -m ipykernel install \
--user --name myderecho_py \
--display-name "Python (myderecho_py)"
````

Test it with some math.

````python
import numpy as np 
 
print("NumPy version:", np.__version__) 
print("2 + 3 =", 2 + 3) 
````

### FEDS Kernel (feds_py39)

Now that we have our own kernel (myderecho_py) that we can adjust and customize to our liking:

Let's set up the official **FEDS kernel** that we will be using for our work.  
It is very similar to our previous example, but we are using Mamba (C++, a faster Conda)     

What are the modules that we are using for the **feds_py39** kernel?  

```module load conda``` loads the Conda environment, very important    
```conda install -n base -c conda-forge mamba``` installs mamba  
```mamba create -n feds_py39 -c conda-forge``` creates environment with conda-forge repository  
  ```python=3.9``` using Python 3.9, we notice all **version numbers** are present, this is for project consistency    
  ```numpy=1.21```   math, handles arrays  
  ```pandas=1.3``` Excel for Python and data science, CSV-style rows and columns     
  ```geopandas=0.9``` handles shapes, polygons, and GPS data, to output tables  
  ```xarray=0.18``` labeled interface, handles multi-dimensional arrays, N-Dimensions    
  ```scipy=1.7```  more math, integration, optimization, stats    
  ```shapely=1.8```    geometric shapes, calculates coordinates area of this specific region    
  ```pyproj=3.2```   calculates coordinates, translator for Coordinate Reference Systems   
  ```gdal=3.4``` engine for reading and writing spatial data formats   
  ```fiona```  specialized for vector data, points and lines     
  ```rasterio```  raster data, satellite and grid imagery    
  ```rtree```  spatial index, quickly find geographic shapes overlap  

We notice the ```ipykernel``` is not here... not yet.  


````python
# copy and paste in terminal 
module load conda 
conda install -n base -c conda-forge mamba
mamba create -n feds_py39 -c conda-forge \
  python=3.9 \
  numpy=1.21 \
  pandas=1.3 \
  geopandas=0.9 \
  xarray=0.18 \
  scipy=1.7 \
  shapely=1.8 \
  pyproj=3.2 \
  gdal=3.4 \
  fiona \
  rasterio \
  rtree
````

Activate from the environment folder we created in the [steps above](#kernelfolder).

````python
mamba info --envs
conda activate /glade/work/$USER/condas/feds_py39
````

Register Jupyter kernel so it is listed and available. 

````python
python -m ipykernel install --user \
  --name feds_py39 \
  --display-name "Python (FEDS py39)"
````

Verify our imports.

````python
python - << 'EOF'
import numpy, pandas, geopandas, xarray, scipy, shapely, pyproj
from osgeo import gdal
print("Environment OK")
EOF
````
