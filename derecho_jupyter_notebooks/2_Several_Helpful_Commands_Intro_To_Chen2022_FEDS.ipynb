## Notebook (.ipynb) Bash Magic and !Bang

**Bash %Magic** commands are used to do bash terminal stuff.

Using a **```!```** (bang) will run in a subshell (a temporary separate process).  
So using **```!cd ~```**  will not work.   
Working with Bash Magic, it is better do **```%cd ~```**  

**IF** done individually a **%Magic** command **WILL** work in the next cell.

Do **```%%bash```** at the top of a cell to use Bash Magic in the entire cell.  
**BUT** this only applies to things inside the cell.  

These commands, if done in a ```%%bash``` cell, **will not** be present in the next cell.

## grep - The Ctrl+F of Terminal

```grep "word"```	Basic Search	Finds the exact string "word" in a file or output.  

```grep -i "word"```	Ignore Case	Finds "Word", "WORD", or "word" automatically.  

```grep -v "word"```	Invert Match	Shows every line except the ones containing "word".  

```grep -r "word" .```	Recursive	Searches for "word" in every file in the current folder.  

```grep -l "word" *```	List Files	Only shows the filenames that contain "word", not the text.  

```grep -n "word"```	Line Numbers	Shows the line number in the file where the match was found.  

```grep -c "word"```	Count	Just tells you how many times the word appears.  

```grep -A 3 "word"```	After Context	Shows the match plus the 3 lines after it.    

```grep -B 2 "word"```	Before Context	Shows the match plus the 2 lines before it.    

```grep -E "a|b"```	Extended (OR)	Finds lines containing either "a" OR "b".  

```grep -P "\d+"```	Perl Regex	Uses advanced patterns like \d for digits.  

```grep -F ".*"```	Fixed String	Treats .* as literal dots and stars, not regex.

## The 'screen' Command - A Virtual Environment  

When working on big projects, we will often need to run multiple process.   
We can (and will) open multiple terminal windows to work, but when left inactive these windows will timeout and you will be logged out.    

The screen command, when called, will create virtual copy of your current environment, this known as a terminal multiplexer.   
With a screen, commands can run while you work on something else.   
Most importantly, if you get kicked off or close your current process window the task will continue.       

Screens will run in the background for long processes, but are also subject to server timeout when idle and will be terminated.  
Once screen is called, you may run tasks or execute code 

```screen -S <name>```   name your screens   
```screen -ls```   list screens.   
Press the keys: **Ctrl a** and then **Ctrl d**, one after the other, to detach from screen and do other work.   
```screen - r <name>```   a screen from the -ls list that you want to reattach to and work from.  
```screen - d <name>```   detach from a screen.   
Press **Ctrl a** and then **?** for screen help list.

## Stats -About This Notebook

````python
import json
import os
import re
from datetime import datetime
from collections import Counter

# ==========================================
#  CONFIGURATION SECTION
# ==========================================

# 1. Path to the notebook you want to analyze
NOTEBOOK_PATH = r"C:\Users\User\Documents\A_JCH_01.30.26\TMCC_MAIN\TMCC_SPRING_SEMESTER_2026\DRI_DATA_Internship\DRI _ Fire _Immergant\Chen_2022\Intro_To_Chen2022_FEDS.ipynb"

# 2. Keywords Configuration
TOP_N_WORDS = 5  # Change this to show more/fewer top words

CUSTOM_IGNORE_WORDS = ["variable", "result", "analysis", "data"] # Add your own words here

EXTERNAL_IGNORE_FILE = None # Path to a .txt file with one word per line (or None)

# 3. Output Configuration 
# Options: 'MARKDOWN' Table (for Jupyter/Colab/VS Code) or 'TEXT' (for Terminal/CLI)
OUTPUT_MODE = 'MARKDOWN' 
# OUTPUT_MODE = 'TEXT'

# ==========================================
# INTERNAL LOGIC 
# ==========================================

def get_stop_words():
    # Base noise, ignore word list
    base_stops = {
        'the', 'and', 'this', 'that', 'with', 'from', 'for', 'was', 'were', 'ipynb', 'output',
        'import', 'print', 'none', 'true', 'false', 'self', 'return', 'name', 'type', 'cell',
        'data', 'list', 'into', 'been', 'have', 'which', 'here', 'also', 'using', 'used',
        'results', 'shown', 'figure', 'table', 'code', 'running', 'index', 'value', 'file'
    }
    
    # Add user-defined words
    base_stops.update([w.lower() for w in CUSTOM_IGNORE_WORDS])
    
    # Load from external file if provided
    if EXTERNAL_IGNORE_FILE and os.path.exists(EXTERNAL_IGNORE_FILE):
        try:
            with open(EXTERNAL_IGNORE_FILE, 'r') as f:
                file_words = f.read().split()
                base_stops.update([w.lower() for w in file_words])
        except Exception as e:
            print(f"Warning: Could not read ignore file: {e}")
            
    return base_stops

def analyze_notebook():
    if not os.path.exists(NOTEBOOK_PATH):
        print(f"Error: File not found at {NOTEBOOK_PATH}")
        return

    with open(NOTEBOOK_PATH, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # 1. Global Setup
    nb_metadata = data.get('metadata', {})
    default_lang = nb_metadata.get('language_info', {}).get('name', 'python').lower()
    LANG_MAP = {
        'python': 'Python', 'python3': 'Python', 'ipython': 'Python',
        'bash': 'Shell', 'sh': 'Shell', 'shell': 'Shell', 'shellscript': 'Shell',
        'zsh': 'Shell', 'batch': 'Shell', 'powershell': 'Shell', 'cmd': 'Shell'
    }

    file_stat = os.stat(NOTEBOOK_PATH)
    file_size = f"{file_stat.st_size / 1024:.2f} KB"
    file_age = datetime.fromtimestamp(file_stat.st_mtime).strftime('%Y-%m-%d %H:%M:%S')

    stats = {
        "code": {"cells": 0, "lines": 0, "chars": 0, "words": 0, "langs": Counter(), "funcs": 0, "vars": 0},
        "md": {"cells": 0, "lines": 0, "chars": 0, "words": 0, "headings": Counter(), "pure_md": 0},
        "all_words": []
    }
    
    stop_words = get_stop_words()

    # 2. Process Cells
    for cell in data.get('cells', []):
        cell_type = cell.get('cell_type')
        source = cell.get('source', [])
        lines = source if isinstance(source, list) else source.splitlines()
        content = "".join(lines)
        clean_content = content.strip()
        
        # Word Analytics (4+ chars only)
        words = re.findall(r'\b[a-zA-Z]{4,}\b', content.lower())
        stats["all_words"].extend([w for w in words if w not in stop_words])

        if cell_type == 'code':
            stats["code"]["cells"] += 1
            stats["code"]["lines"] += len([l for l in lines if l.strip()])
            stats["code"]["chars"] += len(content)
            stats["code"]["words"] += len(words)
            
            # Hierarchical Language Detection
            meta_lang = cell.get('metadata', {}).get('vscode', {}).get('languageId')
            if not meta_lang:
                meta_lang = cell.get('metadata', {}).get('kernelspec', {}).get('language')
            
            magic_match = re.match(r'^%%(\w+)', clean_content)
            is_bang = clean_content.startswith('!')

            if magic_match: detected = magic_match.group(1).lower()
            elif is_bang: detected = 'shell'
            elif meta_lang: detected = meta_lang.lower()
            else: detected = default_lang

            final_lang = LANG_MAP.get(detected, detected.capitalize())
            stats["code"]["langs"][final_lang] += 1

            # Complexity Stats
            stats["code"]["funcs"] += len(re.findall(r'^def\s+\w+\(', content, re.MULTILINE))
            stats["code"]["vars"] += len(re.findall(r'^\s*[a-zA-Z_]\w*(?=\s*=)', content, re.MULTILINE))

        elif cell_type == 'markdown':
            stats["md"]["cells"] += 1
            stats["md"]["chars"] += len(content)
            stats["md"]["words"] += len(words)
            
            has_heading = False
            for line in lines:
                if line.strip().startswith('#'):
                    level_match = re.match(r'^#+', line.strip())
                    if level_match:
                        level = len(level_match.group())
                        stats["md"]["headings"][f"H{level}"] += 1
                        has_heading = True
            
            if not has_heading and clean_content:
                stats["md"]["pure_md"] += 1

    # 3. Final Calculations
    total_h = sum(stats["md"]["headings"].values())
    common_str = ", ".join([f"{w}({c})" for w, c in Counter(stats["all_words"]).most_common(TOP_N_WORDS)])
    h_dist = ", ".join([f"{k}:{v}" for k, v in sorted(stats["md"]["headings"].items())])
    langs_str = ", ".join([f"{k}:{v}" for k, v in stats["code"]["langs"].items()])

    # 4. Generate Report
    report_title = f"Notebook Analysis: {os.path.basename(NOTEBOOK_PATH)}"
    
    if OUTPUT_MODE == 'MARKDOWN':
        from IPython.display import display, Markdown
        report = f"""
### 📊 {report_title}


| Category | Metric | Result |
| :--- | :--- | :--- |
| **System Info** | File Size / Age | {file_size} / {file_age} |
| **Code Analytics** | Total Cells (Lines) | {stats['code']['cells']} ({stats['code']['lines']}) |
| | Avg Word/Char | {stats['code']['words']/max(1, stats['code']['cells']):.1f} / {stats['code']['chars']/max(1, stats['code']['cells']):.1f} |
| | Languages | **{langs_str}** |
| | Complexity | {stats['code']['funcs']} Funcs / {stats['code']['vars']} Vars |
| **Documentation** | Total MD Cells | {stats['md']['cells']} |
| | Info Only (No #) | {stats['md']['pure_md']} |
| | **Total Headings** | **{total_h}** |
| | Heading Dist. | {h_dist} |
| | Avg Word/Char | {stats['md']['words']/max(1, stats['md']['cells']):.1f} / {stats['md']['chars']/max(1, stats['md']['cells']):.1f} |
| **Key Content** | Top {TOP_N_WORDS} Keywords | {common_str} |

**Disclaimer:** *These results are rough estimates based on automated parsing of cell metadata and content patterns.*
"""
        display(Markdown(report))
    else:
        # Plain Text Output for Terminal
        print("-" * 50)
        print(report_title.upper())
        print("-" * 50)
        print(f"File: {file_size} | Modified: {file_age}")
        print(f"Code Cells: {stats['code']['cells']} | Lines: {stats['code']['lines']}")
        print(f"Languages: {langs_str}")
        print(f"Logic: {stats['code']['funcs']} Functions, {stats['code']['vars']} Variables")
        print(f"Total Headings: {total_h} ({h_dist})")
        print(f"Pure Markdown Cells: {stats['md']['pure_md']}")
        print(f"Top {TOP_N_WORDS} Words: {common_str}")
        print("-" * 50)
        print("Disclaimer: These results are rough estimates.")

# Run the analyzer
analyze_notebook()
````
