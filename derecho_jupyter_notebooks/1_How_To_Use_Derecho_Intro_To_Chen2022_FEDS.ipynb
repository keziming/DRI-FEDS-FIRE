## Kernels and Setting The Environment

We will now create a kernel using Conda and Mamba for our project environment.   
We use Conda or Mamba (Mamba is C++ version of Conda) for the Python environment to host the Juptyer Notebook Kernel.

### Install The Modules    

What are the modules that we are using for myderecho_py kernel?  

```conda install -y```  installs the Conda environment

```ipykernel```   connects Conda to our notebook

```scipy```    more math, integration, optimization, stats

## Now, Let’s Setup GitHub!

Linking our notebooks to a repository enables the team to track progress and contribute in meaningful ways.  
  
First, we will establish the connection; then, we will walk through a few workflow examples.

### Github Setup Part 1 - Initial GitHub Setup and 1st Push

--- 1. Setup Directory of Repository ---  
```cd scratch```  # work from your scratch folder  
```git clone https://github.com/keziming/DRI-FEDS-FIRE```  # download the repository, this makes a folder for the files too  
```cd DRI-FEDS-FIRE```  # cd to enter the correct repository folder  
```pwd -P```  # print working directory: confirm you are inside 'DRI-FEDS-FIRE'  
```git```  # test to ensure you are in the right place and have git working  

--- 2. Configure Your Identity ---  
--- Without this, git commit will fail on a new server---  
```git config --global user.name "jhana"```  
```git config --global user.email "jason_hana@mail.tmcc.edu"```  

--- 3. Setup SSH Keys for GitHub Access ---  
```ssh-keygen -t ed25519 -C jason_hana@mail.tmcc.edu```  # generate a new SSH key (press enter for all prompts)  
```cat ~/.ssh/id_ed25519.pub```  # copy the output starting with 'ssh-ed25519...email' to your GitHub Settings -> SSH Keys create new SSH key, title, paste.   
```ssh -T git@github.com```  # test connection: type 'yes' if asked. You want to see "Hi jhana!"  
```git remote set-url origin git@github.com:keziming/DRI-FEDS-FIRE.git```  # switch traffic to SSH  

--- 4. Branching and Navigation ---  
```git fetch```  # download latest updates from the remote without merging  
```git checkout -b nameofbranch```  # creates and switches to your first personal branch named nameofbranch  
```git branch```  # shows names of your local branches ('q' to exit)  
 
--- 5. Creating and Staging Files ---  
```echo "words in a file" > words.txt```  # create a test file  
```ls```  # verify the file exists in your directory  
```git add words.txt```  # stage the file to your branch so Git prepares to track it  

--- 6. Committing and Pushing ---  
```git commit -m "this is the .txt file I’ve added"```  # save the snapshot locally  
```git push -u origin nameofbranch```  # upload your branch and commits to GitHub for the first time  
--- Paste the link from the terminal into a browser to create the Pull Request (PR) ---  

--- 7. Cleanup ---  
```git checkout main```  # switch back to neutral ground  
```ls```  # notice your new files aren't here because they belong to 'nameofbranch'  
```git pull```  # ensure main is up to date  

### Github Setup Part 2 - GitHub Daily Workflow Example

After completing Part 1 above, we will continue to:   
A. Edit the words.txt and   
B. Add a folder for the notebooks

--- 1. Navigate and Update ---  
login to derecho server   
```cd DRI-FEDS-FIRE```  # cd into the correct repository folder  
```git checkout main```  # switch to the stable 'main' branch to get the latest "official" version  
```git pull```  # download the latest 'words.txt' and other team updates  
```ls```  # verify you see the latest files  
  
--- 2. Create Workspace A ---  
Always create a new branch for new work. A branch is a duplicate of the whole project.   
Branches are like paper plates. Name them task-specific like “fix_words_txt”.  
Once completed/commited don’t use them again, delete them. And NEVER work directly on the main branch!   
  
```git checkout -b fix_words_txt```  # create and switch to a branch specifically for the text fix  
```git branch -d nameofbranch```  # optional: delete your OLD branch from Part 1 to keep things clean  
  
--- 3. Edit the File ---  
```cat words.txt```  # read the file  
```echo "I have seen this words.txt file you made, great job" >> words.txt```  # append text (>> adds, > deletes!)  
```cat words.txt```  # verify the text was added correctly  
  
--- 4. Stage and Push Task A ---  
```git status```  # Check status: 'words.txt' should be red (modified)  
```git add words.txt```  # stage the edited text file (turns it green)  
```git commit -m "Updated words.txt with feedback"```  # save snapshot for Task A  
```git push -u origin fix_words_txt```  # send Branch 1 to GitHub  
CLICK THE LINK in the terminal to open GitHub and create the Pull Request for the manager.  
  
--- 5. Return to Main (CRITICAL STEP) ---  
We must go back to main so Task B doesn't accidentally include the work from Task A!  
```git checkout main```  # switch back to main  
```git pull```  # check just in case anyone pushed while we were doing Task A  
  
--- 6. Create Workspace B ---  
```git checkout -b jhana_ipynb_folder```  # create a separate branch for your folder upload  
  
--- 7. Add the Folder and Files ---  
```pwd -P```  # check if you are in the correct repo directory   
```ls```  # verify you see the latest files  
```mkdir derecho_jupyter_notebooks```  # create your new repo folder  
Now, copy your notebooks into this folder. Example command below:  
```cp -r /glade/u/home/jhana/derecho_jupyter/. derecho_jupyter_notebooks/```  # copy your existing work into the repo folder. You may need to use the absolute path of the file location.  
```ls -F derecho_jupyter_notebooks/```  # verify your files are inside the new folder   
  
--- 8. Stage and Push Task B ---  
```git status```  # 'derecho_jupyter_notebooks/' is untracked  
```git add derecho_jupyter_notebooks/```  # stage the ENTIRE folder and its contents (turns it green)  
```git commit -m "Added jason’s notebooks folder"```  # save snapshot for Task B  
```git push -u origin jhana_ipynb_folder```  # send Branch 2 to GitHub  
CLICK THE NEW LINK in the terminal to create a separate PR for the folder.  
  
--- 9. Verification ---  
```git checkout main```  # switch away from your work  
```ls```  # 'derecho_jupyter_notebooks' folder should vanish (it's safe in Branch 2)  
```cat words.txt```  # your new text line should vanish (it's safe in Branch 1)  
  
--- 10. Check your specific work ---  
```git checkout fix_words_txt```  # go to Branch 1  
```cat words.txt```  # text is back!  
```git checkout jhana_ipynb_folder```  # go to Branch 2  
```ls derecho_jupyter_notebooks```  # folder is back!
